```python
import pandas as pd
import numpy as np
```

```python
#Reading CSV file
df = pd.read_csv('spreadspoke_scores.csv')
```

```python
#Creating new column 'score differential' derived from home score minus away score
df['score_differential'] = df['score_home']-df['score_away']
```
